# Справочник типичных комментариев ревьюера

Соберите папку с `.ipynb` (много ноутбуков). Скрипт кластеризует комментарии по **секции** и **цвету Bootstrap-алерта** (`danger` / `warning` / `success`).

**Не ground truth:** справочник отражает частые формулировки, не рубрику и не официальные критерии.

Для метода `tfidf_kmeans` в Colab: `!pip install scikit-learn`

```python
from pathlib import Path
import json
import sys

# Укажите корень репозитория review-assistant (где лежит app/)
REPO = Path("review_assistant_repo")
sys.path.insert(0, str(REPO))

from app.retrieval.comment_catalog import build_catalog, catalog_to_markdown, write_catalog_json

UPLOAD_DIR = Path("review_ipynb")  # положите сюда ноутбуки
PROJECT_SLUG = "colab_catalog"
METHOD = "heuristic"  # или "tfidf_kmeans"

paths = sorted(UPLOAD_DIR.rglob("*.ipynb"))
assert paths, f"Нет .ipynb в {UPLOAD_DIR}"

cat = build_catalog(
    ipynb_paths=paths,
    source_project=PROJECT_SLUG,
    method=METHOD,
    sim_threshold=0.55,
    include_student=False,
)
print(json.dumps(cat["meta"], ensure_ascii=False, indent=2))
write_catalog_json(cat, Path("reviewer_comment_catalog.json"))
Path("reviewer_comment_catalog.md").write_text(catalog_to_markdown(cat), encoding="utf-8")
print("Готово: reviewer_comment_catalog.json, reviewer_comment_catalog.md")
```
